# multi output models experimental

```python
import pandas as pd
import numpy as np

# Initial dataset
data = {
    'N': [21.861871, 95.271719, 69.762452],
    'P': [45.680576, 38.018484, 75.283525],
    'K': [19.686331, 32.593346, 65.065134],
    'label': [
        'peanut,cowpeas,peas,corn',
        'cacao,rice,watermelon',
        'soybeans,rice,banana'
    ]
}

# Create a DataFrame from the initial dataset
initial_df = pd.DataFrame(data)

# Number of dummy datasets to generate
num_datasets = 1000

# Function to generate dummy datasets
def generate_dummy_dataset(initial_df):
    # Create a copy of the initial DataFrame
    dummy_df = initial_df.copy()
    
    # Add random noise to N, P, K values
    noise = np.random.normal(loc=0, scale=5, size=(len(dummy_df), 3))
    dummy_df[['N', 'P', 'K']] += noise
    
    # Modify labels slightly
    for i in range(len(dummy_df)):
        labels = dummy_df.loc[i, 'label'].split(',')
        num_labels = min(3, len(labels))  # Choose a maximum of three labels
        labels = np.random.choice(labels, size=num_labels, replace=False)  # Randomly select labels without replacement
        dummy_df.at[i, 'label'] = ','.join(labels)
    
    return dummy_df

# Generate multiple dummy datasets
dummy_datasets = [generate_dummy_dataset(initial_df) for _ in range(num_datasets)]

# Concatenate all dummy datasets into one DataFrame
combined_df = pd.concat(dummy_datasets, ignore_index=True)

# Save the combined DataFrame to a CSV file
combined_df.to_csv("combined_dummy_datasets.csv", index=False)

print("Combined dummy datasets exported to combined_dummy_datasets.csv")
```

# DTC

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.multioutput import MultiOutputClassifier

# Load the combined dataset from the CSV file
combined_df = pd.read_csv('DTCDummyData.csv')

# Split the data into features (N, P, K) and labels
X = combined_df[['N', 'P', 'K']]
y = combined_df['label'].apply(lambda x: x.split(','))

# Convert labels to binary representation using MultiLabelBinarizer
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
y_binary = mlb.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# Create a multi-output decision tree classifier
dt_classifier = DecisionTreeClassifier(criterion = 'entropy',random_state = 42, max_depth =20)

multi_output_classifier = MultiOutputClassifier(dt_classifier, n_jobs=1)

# Fit the classifier to the training data
multi_output_classifier.fit(X_train, y_train)

# Evaluate the classifier
train_score = multi_output_classifier.score(X_train, y_train) * 100
test_score = multi_output_classifier.score(X_test, y_test) * 100

print(f"Training Accuracy: {train_score:.2f}")
print(f"Testing Accuracy: {test_score:.2f}")
```

```python
import joblib

# Save the model to a file
joblib.dump(multi_output_classifier, 'multi_output_DTC.joblib')
```

```python
import numpy as np

# Predict on the test set
predicted_labels_binary = multi_output_classifier.predict([[90,40,50]])

# Convert the binary predictions back to crop names
predicted_labels = mlb.inverse_transform(predicted_labels_binary)

# Join the predicted crop names for each data point into a single string
predicted_crop_names = np.array([', '.join(crops) for crops in predicted_labels])

# Display the predicted crop names
print("Predicted Crop Names:")
print(predicted_crop_names)
```

# RFC

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble  import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier

# Load the combined dataset from the CSV file
combined_df = pd.read_csv('DTCDummyData.csv')

# Split the data into features (N, P, K) and labels
X = combined_df[['N', 'P', 'K']]
y = combined_df['label'].apply(lambda x: x.split(','))

# Convert labels to binary representation using MultiLabelBinarizer
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
y_binary = mlb.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Create a multi-output decision tree classifier
rf_classifier = RandomForestClassifier(criterion = 'entropy',random_state = 42, max_depth =20)

multi_output_classifier = MultiOutputClassifier(rf_classifier, n_jobs=-1)

# Fit the classifier to the training data
multi_output_classifier.fit(X_train, y_train)

# Evaluate the classifier
train_score = multi_output_classifier.score(X_train, y_train) * 100
test_score = multi_output_classifier.score(X_test, y_test) * 100

print(f"Training Accuracy: {train_score:.2f}")
print(f"Testing Accuracy: {test_score:.2f}")
```
